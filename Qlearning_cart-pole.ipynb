```python
#Qテーブル作成
def bins(clip_min, clip_max, num):
    return np.linspace(clip_min, clip_max, num + 1)[1:-1]
def digitize_state(observation,num):
  observation_low = env.observation_space.low
  observation_high = env.observation_space.high
  digitize_list = [np.digitize(observation[0], bins=bins(-2.4,2.4,num)),
                 np.digitize(observation[1], bins=bins(observation_low[1],observation_high[1], num)),
                 np.digitize(observation[2], bins=bins(observation_low[2],observation_high[2], num)),
                 np.digitize(observation[3], bins=bins(observation_low[3],observation_high[3], num))]

  return  sum([x * (num**i) for i, x in enumerate(digitize_list)])
```

```python
def update_q_table(env,next_observation,observation,action,num,reward):
  alpha = 0.2 #学習率
  gamma = 0.9#時間割引率
  #分割数
  num = 6
#行動後の状態の行動価値
  next_state = digitize_state(next_observation,num)
  next_max_q = q_table[next_state][action]
  #行動前の状態の行動価値
  current_state = digitize_state(observation,num)
  current_max_q = q_table[current_state][action]

  #行動価値関数の更新
  q_table[current_state][action] = current_max_q + alpha * (reward + gamma * next_max_q - current_max_q)

  return q_table,next_state
```

```python
def get_action(q_table,observation,env):
  epsilon = 0.002
  if np.random.uniform(0,1) > epsilon:
    state = digitize_state(observation,6)
    action = np.argmax(q_table[state])
  else:
    action = np.random.choice([0,1])
  return action
```
